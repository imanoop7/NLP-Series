```python
sentence = '''The Effiel Tower is a wrought-iron lattice tower on the Champ de Mars in Paris, France.
            It is named after the engineer Gustave Eiffel, whose company designed and built the tower. 
            Constructed in 1887-1889 as the entrance to the 1889 World's Fair, it was initially criticized by 
            some of France's leading artists and intellectuals for its design, but it has become a global cultural
            icon of France and one of the most recognisable structures in the world. 
            The Eiffel Tower is the most-visited paid monument in the world; 6.91 million people ascended it 
            in 2015. The tower is 324 metres (1,063 ft) tall, about the same height as an 81-storey building, 
            and the tallest structure in Paris. Its base is square, measuring 125 metres (410 ft) on each side.
            During its construction, the Eiffel Tower surpassed the Washington Monument to become the tallest.

'''
```

```python
import nltk
words = nltk.word_tokenize(sentence)
```

```python
words
```

```python
tag_elements = nltk.pos_tag(words)
```

```python
tag_elements
```
