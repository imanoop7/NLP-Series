```python
#Bag of words is a simple and commonly used model for text classification.
#It involves two things:
#1. A vocabulary of known words.
#2. A measure of the presence of known words.
#It is called a “bag” of words, because any information about the order or structure of words in the document is discarded. 
#The model is only concerned with whether known words occur in the document, not where in the document.
#The intuition is that documents are similar if they have similar content. Further, that from the content alone we can learn something about the meaning of the document.
```

```python
# how Bag of Words works:
#1. Collect Data: You can use any data, but it must be in a format that you can extract words from.
#2 Stemming or lemmatization: You can use stemming or lemmatization to reduce words to their root form.
#3. Remove stop words: You can remove common words called stop words.
#4. Construct a vocabulary: from the unique words in your data.
#5. Encode the data: You can encode your documents using the vocabulary.
#6. Vectorize the data: You can convert the encoded documents into a sparse matrix.
```

```python
import pandas as pd
messages = pd.read_csv('SMSSpamCollection.txt', sep='\t', names=["label", "message"])
```

```python
messages
```

```python
# data cleaning and preprocessing 
# removing special characters and stop words
# stemming
```

```python
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
stemmer = PorterStemmer()
```

```python
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['message'][i])
    review = review.lower()
    review = review.split()
    
    review = [stemmer.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
# lemetization
corpus = []
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['message'][i])
    review = review.lower()
    review = review.split()
    
    review = [lemmatizer.lemmatize(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500)
```

```python
X = cv.fit_transform(corpus).toarray()
```
